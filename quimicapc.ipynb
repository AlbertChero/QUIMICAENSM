LEY DE BOYLE

```python
ley_boyle=pd.read_csv('dboyle.csv')
ley_boyle
```

```python
plt.scatter(1/ley_boyle["x"],ley_boyle["y"])
```

```python
def RegLin (LX, LY):
  sumX = 0
  sumY = 0
  sumXY = 0
  sumXcuadrado = 0
  for elemento in LX:
    sumX += elemento
    sumXcuadrado = elemento**2
  for elemento in LY:
    sumY += elemento
  i = 0
  for elemento in LX:
    sumXY = LX[i]*LY[i]
    i+=1
  p = i

  m = (((p*sumXY)-(sumX*sumY))/((p*sumXcuadrado)-(sumX*sumX)))
  b = (((sumXcuadrado*sumY)-(sumX*sumXY))/((p*sumXcuadrado)-(sumX*sumX)))
  return f"{round(m,1)}x + {round(b,1)}"
print("p =", RegLin(ley_boyle["x"], ley_boyle["y"]))
```
